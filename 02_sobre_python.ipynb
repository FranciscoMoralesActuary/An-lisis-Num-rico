### **1.1 Mutabilidad de objetos**

Todos los tipos de datos en Python están representados por objetos o por relaciones entre objetos. Cada objeto tiene una identidad, un tipo y un valor:

*   La identidad de un objeto nunca cambia una vez que ha sido creado; puedes pensar en ella como la dirección del objeto en la memoria.
*   El tipo de un objeto determina las operaciones que admite y define los posibles valores que puede tener.
*   El valor es el dato específico asignado a una variable. Los objetos cuyo valor puede cambiar se llaman mutables, mientras que aquellos cuyo valor no puede modificarse una vez creados se llaman inmutables.

La mutabilidad de un objeto está determinada por su tipo; por ejemplo, los números, las cadenas de caracteres y las tuplas son inmutables, mientras que los diccionarios y las listas son mutables.






```python
# Ejemplo inmutabilidad

x = 1
print("La identidad de x es", id(x), ", y su tipo es", type(x), ", y su valor es", x)

x = 2
print("La identidad de x es", id(x), ", y su tipo es", type(x), ", y su valor es", x)
```

```python
# Otro ejemplo de inmutabilidad

y = (1,2,3)
print("La identidad de y es", id(y), ", y su tipo es", type(y), ", y su valor es", y)

y.append(4)
print("La identidad de y es", id(y), ", y su tipo es", type(y), ", y su valor es", y)
```

```python
# Ejemplo de mutabilidad

y = [1,2,3]
print("La identidad de y es", id(y), ", y su tipo es", type(y), ", y su valor es", y)

y.append(4)
print("La identidad de y es", id(y), ", y su tipo es", type(y), ", y su valor es", y)
```

## **2. Control if y sus variantes**

Tenemos los controles `if` o `if else` que no sirve para tomar una decisión.



```python
# Control básico IF

# Estructura:
#
# if CONDICION:
#    accion si CONDICION es verdadero
#

# Definimos unas variables a usar:
x = 2
y = 3

# Ejemplo
if x > y:
  print(x," es menor que ", y)

```

```python
# Control IF - ELSE (CUIDADO CON EL IDENTADO)

# Estructura:
#
# if CONDICION:
#    accion si CONDICION es verdadero
# else:
#    accion si CONDICION es falso
#

# Ejemplo

if x**2 < 0:
  print("WTF!!!")
else:
  print("Todo bien bandita")
```

```python
# Control IF - ELIF - ELSE (CUIDADO CON EL IDENTADO)

# Estructura:
#
# if CONDICION1:
#    accion si CONDICION1 es verdadero
# elif CONDICION2:
#    accion si CONDICION2 es verdadero
# else:
#    accion si las dos anteriores no pasan

# Ejemplo
if x**2 < 0:
  print("WTF!!!")
elif x**2 == 0:
  print("Solo si",x,"es el cero. ¡Ah pues si lo es!")
else:
  print("Todo bien, es un real")
```

## **3. Ciclo for**

`for` nos sirve para hacer una actividad en repetidas ocaciones por un número de veces que fijamos.

Este número de veces que hacemos una acción tiene que ir indexado por algo, aquí trabajamos con listas o tuplas.



```python
# Estructura
# for variable in lista_de_indices:
#    la_acción
#


#Como creamos listas o tuplas

#Así creamos listas
indices_lista = [1,2,3,4]
#Así tuplas
indices_tupla = (1,2,3,4)

# Ciclo sobre lista
for k in indices_lista:
  print("vamos en el ciclio",k, "sobre la lista")

# Ciclo sobre tupla
for k in indices_tupla:
  print("vamos en el ciclio",k, "sobre la tupla")
```

## **4. Ciclo while**

`while` nos sirve para hacer una actividad en repetidas ocaciones mientras la condicion sea cierta, continuará ejecutandose el ciclo, y solo se detendrá cuando la condición deje de ser verdadera. Hay que tener cuidado con este tipo de ciclos porque se pueden volver infinitos si las condiciones no están bien puestas


```python
# Estructura

# while CONDICION:
#   ACCION o ACCIONES
#

# Ejemplo del while
m = 100
k = 0
while m > 10:
  print("aqui sigo en el ciclo, en el paso",m)
  m = m-k
  k = k+1  # ¡Importante incrementar una unidad en cada iteración!


```

## **5. Controles de flujo**

En la programación, los ciclos suelen emplear dos comandos fundamentales para controlar su flujo: `break` y `continue`.

`break`: Este comando se utiliza cuando se cumple una condición específica dentro del ciclo, `break` permite salir de él de manera inmediata, sin ejecutar más iteraciones.


```python
for num in range(6, 20):
    if num % 5 == 0:
        print(f"El ciclo termino, encontramos un número divisible por 5: {num}")
        break
    print(f"Número actual: {num}")
```

`continue`: A diferencia de `break`, `continue` no termina el ciclo, sino que interrumpe la iteración actual y pasa directamente a la siguiente.

```python
i=0
while i<11:
  i = i+1
  if i % 3 == 0:
    continue
  print(f"Número no divisible por 3: {i}")
```

### **6.1 Funciones Recursivas**

Las funciones recursivas son aquellas funciones que se llaman a ellas mismas en su propia ejecucción. Es importante tener cuidado en este tipo de funciones ya que puede caer en bucles infinitos.

```python
# Ejemplo Fibonacci

# Creamos una función que imprima el i-ésimo valor de la sucesión de Fibonacci.
# Recuerde que la función de Fibonacci es 0,1,1,2,3,5,8,13,21,...

def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5)
```

##  **7. Ejercicio**
Escribir una implementacion de algún algoritmo que pase un número entero $n$ positivo a base 2.

#### **Observación:**
El código abajo es una implementación que se me ocurrió del algoritmo clásico. Es claro que debe haber códigos más eficientes, pero la idea es hacer algo más con lo que tengamos a la mano.

$2 = 1\cdot 2^1+0\cdot 2^0 \Rightarrow [2]_2=10$

$10 = 1\cdot 2^3+0\cdot 2^2+1\cdot 2^1+0\cdot 2^0 \Rightarrow [10]_2=1010$

```python
# Ejercicio

# Aquí puedes modificar el número:
mi_numero = 10

numero_base2 = list()

# Variables auxiliares
aux1 = mi_numero

#Vamos a ciclearlo
while aux1 > 0:
  k = 0
  while 2**k <= aux1:
    print("vamos en el paso ",k)
    k = k+1
  numero_base2.append(k-1)
  aux1 = aux1 - 2**(k-1)
#Veamoslo (solo las potencias)

print("Las potencias de 2 que aparecen son",
      " ".join(map(str,numero_base2)))


#Creamos una variable pa jugar
potencia = numero_base2[0]+1

#Una lista pa' guardar
mi_numero2 = [0]*potencia

#Un ciclo pa rellenar
for j in range(numero_base2[0]+1):
  if j in numero_base2:
    mi_numero2[potencia-1-j]=1

#Y he aquí el número
print("\n El número",mi_numero,"se escribe como",
      "".join(map(str,mi_numero2)), "en base 2,")
print("esto es: (",mi_numero,")_2 = ","".join(map(str,mi_numero2)), sep='')

```
