### Introducción a Matplotlib

Matplotlib es una biblioteca de Python especializada en la visualización de datos. En combinación con el módulo pyplot, facilita la creación de gráficos y representaciones visuales.

```python
# Importamos la librería renombrandola como plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Generamos un DataFrame a partir de un Diccionario
datos = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Región": ["Norte", "Sur", "Este", "Oeste", "Norte", "Sur", "Este", "Oeste", "Norte", "Sur", "Este", "Oeste"],
    "Celulares": [120, 150, 180, 200, 170, 160, 140, 130, 190, 210, 220, 250],
    "Pantallas": [100, 110, 140, 130, 125, 135, 150, 145, 155, 165, 170, 180],
    "Consolas": [200, 190, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300],
    "Bocinas": [300, 310, 320, 330, 340, 350, 360, 370, 380, 390, 400, 410]}

df = pd.DataFrame(datos)
df
```

### Gráfico de barras

A continuación se muestra cómo se crea un gráfico de barras utilizando Matplotlib. Para complementar los gráfico generalmente se utilizan las mismas funciones básicas que nos ayudan a hacer el gráfico más bonito. La estructura principal del código no cambia, solo se modifican ciertos detalles, como el tipo de gráfico y los parámetros asociados.

```python
# Configuramos el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Creamos un gráfico de barras especificando el eje x, eje y, color de las barras y su etiqueta
plt.bar(df["Mes"], df["Celulares"], color='skyblue', label="Celulares")

# Agregamos un título al gráfico con un tamaño de fuente
plt.title("Ventas mensuales de Celulares", fontsize=14)

# Etiquetamos el eje x
plt.xlabel("Mes", fontsize=12)

# Etiquetamos el eje y
plt.ylabel("Ventas", fontsize=12)

# Ajustamos la rotación de las etiquetas en el eje x
plt.xticks(rotation=45)

# Mostramos la leyenda para identificar los datos representados en el gráfico
plt.legend()

# Agregamos una cuadrícula horizontal con un estilo punteado y una transparencia
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostramos el gráfico
plt.show()
```

### Gráfico de Pastel

```python
# Como se comento generalmente se complementa el gráfico con las mismas funciones antes mencionadas simplemente cambia el tipo de gráfico y sus parametros
# (al final hay una tabla que explica cada tipo de gráfico y sus parametros)

# Agrupamos los valores de la suma de cada producto por región
v3 = df.groupby("Región")[["Celulares", "Pantallas", "Consolas", "Bocinas"]].sum().sum(axis=1)

plt.figure(figsize=(8, 8))
plt.pie(v3, labels=v3.index, autopct='%1.1f%%', colors=["gold", "skyblue", "lightgreen", "pink"])
plt.title("Distribución de ventas totales por región", fontsize=14)
plt.show()

```

### Gráfico de líneas

```python
plt.figure(figsize=(12, 6))

# Generamos el gráfico de cada producto por mes
for producto in ["Celulares", "Pantallas", "Consolas", "Bocinas"]:
    plt.plot(df["Mes"], df[producto], marker='o', label=producto)

plt.title("Tendencia de ventas mensuales por producto", fontsize=14)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Ventas", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Productos")
plt.grid()
plt.show()
```

### Histograma

```python
plt.hist(np.random.standard_normal(100), bins=50, color='purple')
plt.title('Histograma de una Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

### Tabla de tipos de gráficos

| Tipo de Gráfico                | Parámetro             | Descripción                                                                                        |
|---------------------------------|-----------------------|----------------------------------------------------------------------------------------------------|
| **Gráfico de Barras** (`plt.bar`)      | `height`              | Altura de las barras                                   |
|                                 | `width`               | Ancho de las barras.                                                                                |
|                                 | `color`               | Color de las barras.                                                                                |
|                                 | `label`               | Etiqueta para cada barra, usada para la leyenda.                                                   |
|                                 | `alpha`               | Transparencia de las barras (de 0 a 1).                                                            |
| **Gráfico de Líneas** (`plt.plot`)     | `x`                   | Valores en el eje X.                                                                                |
|                                 | `y`                   | Valores en el eje Y.                                                                                |
|                                 | `color`               | Color de la línea.                                                                                 |
|                                 | `linestyle`           | Estilo de la línea, como `'-'` (sólida), `':'` (punteada), `--` (guiones).                         |
|                                 | `linewidth`           | Grosor de la línea.                                                                                 |
|                                 | `marker`              | Marca de los puntos en la línea, como `'o'`, `'^'`, `'.'`, etc.                                     |
| **Gráfico de Dispersión** (`plt.scatter`)| `x`                   | Valores en el eje X.                                                                                |
|                                 | `y`                   | Valores en el eje Y.                                                                                |
|                                 | `color`               | Color de los puntos.                                                                                |
|                                 | `s`                   | Tamaño de los puntos (en puntos cuadrado).                                                         |
|                                 | `alpha`               | Transparencia de los puntos.                                                                        |
| **Gráfico de Pastel** (`plt.pie`)      | `labels`              | Etiquetas que describen cada sección del pastel.                                                     |
|                                 | `autopct`             | Formato de porcentaje mostrado dentro de cada sección (como `'%1.1f%%'`).                          |
|                                 | `colors`              | Colores de cada sección del gráfico.                                                               
| **Gráfico de Área** (`plt.fill_between`)| `x`                   | Valores en el eje X.                                                                                |
|                                 | `y1`                  | Valores en el eje Y para la primera serie de datos.                                                  |
|                                 | `y2`                  | Valores en el eje Y para la segunda serie de datos (puede ser una línea horizontal).               |
|                                 | `color`               | Color del área.                                                                                   |
| **Histograma** (`plt.hist`)            | `bins`                | Número o límites de los bins (intervalos) del histograma.                                          |
|                                 | `color`               | Color de las barras del histograma.                                                                |
|                                 | `alpha`               | Transparencia de las barras.                                                                       |
|                                 | `label`               | Etiqueta para el histograma, útil para la leyenda.                                                 |
| **Gráfico de Caja** (`plt.boxplot`)    | `data`                | Datos que serán graficados.                                                                        |
|                                 | `vert`                | Si es `True`, la caja será vertical, si es `False`, horizontal.                                    |
|                                 | `notch`               | Si es `True`, se agrega una muesca en la caja para representar la mediana.                         |
|                                 | `patch_artist`        | Si es `True`, rellena la caja con color.                                                           |
