```python
# prompt: programa el metodo de la biseccion

def biseccion(f, a, b, tolerancia=1e-6, iteraciones_maximas=100):
    """
    Encuentra una raíz de la función f en el intervalo [a, b] usando el método de la biseccion.

    Args:
        f: La función cuya raíz se busca.
        a: El extremo izquierdo del intervalo.
        b: El extremo derecho del intervalo.
        tolerancia: La precisión deseada para la raíz.
        iteraciones_maximas: El número máximo de iteraciones permitidas.

    Returns:
        Una tupla (raiz, numero_iteraciones) donde raiz es una aproximación de la raíz de f
        en el intervalo [a, b] y numero_iteraciones es el número de iteraciones realizadas.
        Devuelve None si no se encuentra una raíz dentro de la tolerancia o del numero de iteraciones.
    """

    if f(a) * f(b) >= 0:
        print("La función no cambia de signo en el intervalo dado.")
        return None

    iteracion = 0
    while (b - a) / 2 > tolerancia and iteracion < iteraciones_maximas:
        c = (a + b) / 2
        if f(c) == 0:
            return c, iteracion  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1

    raiz_aproximada = (a + b) / 2
    return raiz_aproximada, iteracion
```

### **Ejercicio 10 (1.17)**

Usando la fórmula cuadrática, calcula la solución positiva de la ecuación dada. Si tiene alguna complicación al calcular la solución, explique cómo resolvió el problema.

(a) $x^2+10^8x− \frac{1}{4} = 0$

(b) $10^{−16}x^2+x− \frac{1}{4} = 0 $

Sea $ax^2 + bx +c = 0$
$$x = \frac{-b\pm \sqrt{b^2 - 4ac}}{2a}$$



```python
import math

def resolver_ecuacion_cuadratica(a, b, c):
  """Resuelve una ecuación cuadrática de la forma ax^2 + bx + c = 0 usando la fórmula general.

  Args:
    a: El coeficiente del término x^2.
    b: El coeficiente del término x.
    c: El término constante.

  Returns:
    Una tupla que contiene las dos soluciones de la ecuación, o None si la ecuación no tiene soluciones reales.
  """

  discriminante = b**2 - 4*a*c

  if discriminante < 0:
    return None  # No hay soluciones reales
  elif discriminante == 0:
    x = -b / (2*a)
    return (x, x)  # Solución única (raíz doble)
  else:
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)
    return (x1, x2)  # Dos soluciones distintas
```

```python
#Ejercicio 2
resolver_ecuacion_cuadratica(10**-16,1,-0.25)
```

```python
# prompt: Me puedes generar un codigo que me de el metodo de la bisección

def biseccion(f, a, b, tolerancia=1e-4, max_iteraciones=100):
  """
  Encuentra una raíz de la función f en el intervalo [a, b] usando el método de la bisección.

  Args:
    f: La función para la cual se busca la raíz.
    a: El extremo izquierdo del intervalo.
    b: El extremo derecho del intervalo.
    tolerancia: La tolerancia para la convergencia.
    max_iteraciones: El número máximo de iteraciones.

  Returns:
    Una tupla (aproximación de la raíz, número de iteraciones) o None si no se encuentra una raíz.
  """

  if f(a) * f(b) >= 0:
    print("La función no cumple con las condiciones para el método de la bisección.")
    return None

  iteraciones = 0
  for _ in range(max_iteraciones):
    c = (a + b) / 2
    if abs(f(c)) < tolerancia:
      return c, iteraciones + 1
    elif f(a) * f(c) < 0:
      b = c
    else:
      a = c
    iteraciones += 1

  print("Se alcanzó el número máximo de iteraciones.")
  return c, iteraciones + 1
```

```python
# prompt: me puedes generar la funcion de tal manera que me vaya poniendo la iteracion, valores de x, a,b y el error absoluto en un dataframe

import pandas as pd

def biseccion(f, a, b, tolerancia=1e-6, max_iteraciones=100):
  """
  Encuentra una raíz de la función f en el intervalo [a, b] usando el método de la bisección.
  Almacena las iteraciones, valores de x, a, b y error absoluto en un DataFrame.

  Args:
    f: La función para la cual se busca la raíz.
    a: El extremo izquierdo del intervalo.
    b: El extremo derecho del intervalo.
    tolerancia: La tolerancia para la convergencia.
    max_iteraciones: El número máximo de iteraciones.

  Returns:
    Un DataFrame con la información de cada iteración y la aproximación de la raíz.
  """

  if f(a) * f(b) >= 0:
    print("La función no cumple con las condiciones para el método de la bisección.")
    return None

  data = []
  for i in range(max_iteraciones):
    c = (a + b) / 2
    error_absoluto = abs(f(c))
    data.append([i + 1, c, a, b, error_absoluto])

    if abs(f(c)) < tolerancia:
      df = pd.DataFrame(data, columns=['Iteracion', 'x', 'a', 'b', 'Error Absoluto'])
      return df, c
    elif f(a) * f(c) < 0:
      b = c
    else:
      a = c

  print("Se alcanzó el número máximo de iteraciones.")
  df = pd.DataFrame(data, columns=['Iteracion', 'x', 'a', 'b', 'Error Absoluto'])
  return df, c

```
